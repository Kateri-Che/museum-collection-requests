#### Задача 1####
- Найти все произведения, связанные с ананасами. 
- Выбрать id тех объектов, которые присутствуют в списке (40007, 156931, 450114, 418070, 436524).

```python
#выполним поиск по слову 'pineapples'
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?q=pineapples')
res = r.json()
```

```python
#Сохраним id полученных объектов в список pineapples_ids
pineapples_ids = res['objectIDs']
pineapples_ids
```

```python
bucket = list()
l = [40007, 156931, 450114, 418070, 436524]
for i in l:
    if i in pineapples_ids:
            bucket.append(i)
    else:
        continue
print(bucket)
```

**Ответ**: при поиске по объектам, связаным с ананасами, были найдены два объекта из списка: 156931, 450114.

#### Задача 3 ####
- Написать цикл, с помощью которого будет собрана информация об объектах, которые были получены в задаче 1. 
- Результаты записать в датафрейм:

 - object_id – id объекта
 - accession_year - год поступления в коллекцию музея
 - title – название
 - artist_name – автор
 - department – департамент
 - object_begin_date – дата (начало)
 - object_end_date – дата (конец)
 - period – название периода
 - object_name – название/категория объекта
 - culture – культура
- К каким культурам относятся объекты? Подсчитать количество объектов, у которых в данном разделе ничего не указано.

```python
# распарсим json через словарь
setdata = pd.DataFrame()
for i in pineapples_ids:
    try:
        r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/objects/' + str(i))
        result = r.json()
        data = {'object_id': [result['objectID']],
                'accession_year': [result['accessionYear']],
                'title': [result['title']], 
                'artist_name': [result['artistDisplayName']], 
                'department': [result['department']], 
                'object_begin_date': [result['objectBeginDate']], 
                'object_end_date': [result['objectEndDate']], 
                'period': [result['period']], 
                'object_name': [result['objectName']], 
                'culture': [result['culture']]}
        df = pd.DataFrame(data)
        setdata = pd.concat([setdata, df], ignore_index=True)
    except:
        continue
setdata
```

```python
setdata['culture'].unique()
```

```python
setdata.query("culture == ''")['culture'].count()
```

**Ответ:** Объекты,связанные с ананасами, относятся к культуре Индии, Испании, Америки, Франции, Филиппин, Китая, Италии, Коромандельского побережья Индии (для рынка Индонезии), народа Фон Королевства Дагомея, этнической группы Вили народа Конго. У 64 объектов в разделе "культура" стоит знак пропуска.

#### Задача 5 ####
- Построить запрос через секцию "Search".
- Выполнить поисковый запрос для департамента "Drawings and Prints" (departmentId = 9) с тегом "mountains" и записать первые 100 id объектов в датафрейм (по аналогии с датафреймом в задаче 4).
- Найти художника, которому принадлежит большинство работ, находящихся на хранении в музее.
- Найти год, в который поступило болшинство объектов в коллекцию музея.

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=9&q=mountains')
mountains = r.json()
```

```python
mountains_id = mountains['objectIDs']
```

```python
setdata_mountains = pd.DataFrame()
for i in mountains_id[:100]:
    try:
        r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/objects/' + str(i))
        result = r.json()
        data = {'object_id': [result['objectID']],
                'accession_year': [result['accessionYear']],
                'title': [result['title']], 
                'artist_name': [result['artistDisplayName']], 
                'department': [result['department']], 
                'object_begin_date': [result['objectBeginDate']], 
                'object_end_date': [result['objectEndDate']], 
                'period': [result['period']], 
                'object_name': [result['objectName']], 
                'culture': [result['culture']]}
        df = pd.DataFrame(data)
        setdata_mountains = pd.concat([setdata_mountains, df], ignore_index=True)
    except:
        continue
setdata_mountains
```

```python
setdata_mountains['artist_name'].value_counts()
```

```python
setdata_mountains['accession_year'].value_counts()
```

**Ответ:** Из 100 музейных объектов, относящихся к департаменту "Drawings and Prints" и связанных с горами, большинство работ принадлежит художнику Уильяму Тёрнеру. Самое большое количество объектов из 100 поступило в коллекцию музея в 2001, 2006 и 2012.
